# Notebook for plotting geocentric axial dipole model properties

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
label_lats = [-89.9, -60, -30, 0, 30, 60, 89.9]
```

```python
# now let's make a I vs. lat plot
plt.figure(figsize=(6,6))
lats = np.linspace(-90, 90, 100)
incs = np.degrees(np.arctan(2 * np.tan(np.radians(lats))))

label_incs = np.degrees(np.arctan(2 * np.tan(np.radians(label_lats))))

plt.plot(lats, incs)
plt.scatter(label_lats, label_incs, color='k')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(-90, 90)
plt.ylim(-90, 90)
plt.xlabel('Latitude ($^\circ$)', fontsize=14)
plt.ylabel('Inclination ($^\circ$)', fontsize=14)

plt.savefig('../figure/GAD_inc_vs_lat.pdf', bbox_inches='tight')
plt.show()
```

```python
# now let's make a surface field intensity vs. lat plot
plt.figure(figsize=(6,6))
lats = np.linspace(-90, 90, 100)
ints = np.sqrt(3*np.sin(np.radians(lats))**2 + 1)

label_ints = np.sqrt(3*np.sin(np.radians(np.array(label_lats)))**2 + 1)

plt.plot(lats, ints)
plt.scatter(label_lats, label_ints, color='k')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(-90, 90)
plt.ylim(1, 2)
plt.xlabel('Latitude ($^\circ$)', fontsize=14)
plt.ylabel('Geomagnetic field intensity\n(normalized)', fontsize=14)
plt.savefig('../figure/GAD_int_vs_lat.pdf', bbox_inches='tight')
plt.show()
```
