# Notebook for making paleointensity experiment illustrations

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
plt.figure(figsize=(6,6))

NRMs = np.linspace(1, 0, 9)
TRMs = np.linspace(0, 1, 9)

n = np.arange(2, 10, 2)

for i in n:
    plt.plot([NRMs[i-2], NRMs[i]], [TRMs[i-2], TRMs[i-2]], color='grey', linestyle='--')
    plt.plot([NRMs[i], NRMs[i]], [TRMs[i], TRMs[i-2]], color='grey', linestyle='--')

plt.scatter(NRMs[n], TRMs[n], color='none', edgecolor='grey', ls='--', marker='^', s=300)

plt.plot(NRMs, TRMs, color='black')

for i, (NRM, TRM) in enumerate(zip(NRMs, TRMs)):
    if i % 2 == 0:
        plt.scatter(NRM, TRM, color='black', marker='s')
    else:
        plt.scatter(NRM, TRM, color='black', marker='o')

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)

plt.ylabel('Normalized natural remanence\n(TRM$_{paleo}$)', fontsize=14)
plt.xlabel('Normalized lab remanence\n(TRM$_{lab}$)', fontsize=14)
plt.savefig('../figure/IZZI_protocol.pdf', bbox_inches='tight')
plt.show()
```
